```python
# otimização de regressão para ML
import numpy as np
import matplotlib.pyplot as plt
W = 0;
learning_rate=0.1
T=10
x_train = np.array([-2.0,  0.0, 4.0, 1.0], dtype=float)
y_train = np.array([2.0,  0.0, -4.0, -1.0], dtype=float)
loss_sq = np.zeros(T)
for t in range(0,T):
  for n in range(0,y_train.size):
    score = W*x_train[n]
    residuo = score-y_train[n]
    grad = 2*residuo*x_train[n]
    W=W-learning_rate*grad
  loss_sq[t]=residuo**2  
  print('loss e W=',[loss_sq[t], W])
plt.plot(loss_sq,'k'),plt.xlabel('iteraçao'),plt.ylabel('Loss')
print('W final=',W)
#fazendo os gráficos
plt.rcParams.update({'font.size': 12})
plt.figure()
plt.plot(x_train,x_train*W,'r.')
plt.plot(x_train,y_train,'b-')
plt.legend(['previsões','reta original'])
plt.xlabel('x');plt.ylabel('y')
```

```python
# otimização de classificação para ML
import numpy as np
import matplotlib.pyplot as plt
W = np.array([0, 0])
learning_rate=0.1
T=10
x_train = np.array([[2.0,  0.0],[3.0, 1.0],[3.0, 3.0]], dtype=float)
y_train = np.array([[-1],  [-1], [1]])
loss_lg = np.zeros(T)
for t in range(0,T):
  for n in range(0,y_train.size):
    score = np.dot(W,x_train[n])
    margin = score*y_train[n]
    grad = y_train[n]*(-x_train[n]/(1+np.exp(margin)))
    W=W-learning_rate*grad
    previsao=np.sign(np.dot(W,np.transpose(x_train)))
  loss_lg[t]=np.log(1+np.exp(-margin))
  print('previsão=',previsao)
print('previsao final=',previsao)
print('valores esperados=',np.transpose(y_train))
plt.plot(loss_lg,'k')
plt.xlabel('iteraçao'),plt.ylabel('Loss')
print('W final=',W)
```
